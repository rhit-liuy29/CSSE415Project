```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
df = pd.read_csv(".\Data\creditcard.csv")
df.Class = df.Class.astype('category')
print(df.shape)
df.head(3)
```

```python
df_fraud = df[df.Class==1].sample(n=100)
df_not = df[df.Class==0].sample(n=100)
df_fraud.shape
```

```python
df_not.shape
```

```python
df = df_fraud.merge(df_not, how='outer')
df.head(3)
```

```python
for feature in df.columns[0:30]:
    df.boxplot(column=feature, by='Class')
```
